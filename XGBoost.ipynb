```python
import xgboost as xgb
import pickle as pkl
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = open("train_df.pkl",'rb')
train_df = pkl.load(file)
file = open("test_df.pkl",'rb')
test_df = pkl.load(file)
file = open("closed_store_ids.pkl",'rb')
closed_store_ids = pkl.load(file)
```

```python
train_dic = dict(list(train_df.groupby('Store')))
test_dic     = dict(list(test_df.groupby('Store')))
submission1   = Series()
submission50   = Series()
submission100   = Series()
submission150   = Series()
submission200   = Series()

for i in test_dic:
    store = train_dic[i]
    
    X_train = store.drop(["Sales","Store"],axis=1)
    Y_train = store["Sales"]
    X_test  = test_dic[i].copy()
    
    store_ids = X_test["Id"]
    X_test.drop(["Id","Store"], axis=1,inplace=True)
    
    params = {"objective": "reg:linear",  "max_depth": 50}
    train_xgb = xgb.DMatrix(X_train, Y_train)
    test_xgb  = xgb.DMatrix(X_test)
    
    gbm1 = xgb.train(params, train_xgb, 1)
    gbm50 = xgb.train(params, train_xgb, 50)
    gbm100 = xgb.train(params, train_xgb, 100)
    gbm150 = xgb.train(params, train_xgb, 150)
    gbm200 = xgb.train(params, train_xgb, 200)
    
    Y_pred = gbm1.predict(test_xgb)
    submission1 = submission1.append(Series(Y_pred, index=store_ids))
    Y_pred = gbm50.predict(test_xgb)
    submission50 = submission50.append(Series(Y_pred, index=store_ids))
    Y_pred = gbm100.predict(test_xgb)
    submission100 = submission100.append(Series(Y_pred, index=store_ids))
    Y_pred = gbm150.predict(test_xgb)
    submission150 = submission150.append(Series(Y_pred, index=store_ids))
    Y_pred = gbm200.predict(test_xgb)
    submission200 = submission200.append(Series(Y_pred, index=store_ids))

```

```python
submission1 = submission1.append(Series(0, index=closed_store_ids))
submission50 = submission50.append(Series(0, index=closed_store_ids))
submission100 = submission100.append(Series(0, index=closed_store_ids))
submission150 = submission150.append(Series(0, index=closed_store_ids))
submission200 = submission200.append(Series(0, index=closed_store_ids))
```
